```python
#imports & styling for chart
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

```python
#retrieve data from the csv files and assign them to a variable
newYorkData = pd.read_csv('NewYork.csv')
globalData = pd.read_csv('Global.csv')
```

```python
#plot New York's average temperatures
plt.plot(newYorkData.year, newYorkData.avg_temp);
#plot Global average temperatures
plt.plot(globalData.year, globalData.avg_temp);
#legend and labels
plt.legend(['New York', 'Global']);
plt.ylabel('Average Temperature');
plt.xlabel('Year');
plt.show();
```

```python
#plot New York's average temperatures with a 7 day rolling average
plt.plot(newYorkData.year, newYorkData.avg_temp.rolling(7).mean())
plt.plot(globalData.year, globalData.avg_temp.rolling(7).mean())
#legend and labels
plt.legend(['New York', 'Global']);
plt.ylabel('Average Temperature');
plt.xlabel('Year');
plt.show();
```
